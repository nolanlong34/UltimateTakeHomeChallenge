```python
import pandas as pd
import seaborn as sns
```

```python
#imports the data
logins = pd.read_json('logins.json')
```

```python
#sorts the data into 15 minute intervals
logins['login_time'] = pd.to_datetime(logins['login_time'])
logins['count'] = 1
logins = logins.set_index('login_time')
logins = logins.resample('15T').sum()
```

```python
#creates a column for the day of the week
logins['day'] = logins.index.dayofweek
```

```python
#creates a chart of logins by day of the week using seaborn
barplot  = sns.barplot(x='day', y='count', data=logins)
barplot.set_xticks(range(7))
barplot.set_xticklabels(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
barplot.set_title('Logins by Day of the Week')
```

```python
#plots average logins by time of day
logins['time'] = logins.index.time
average_time_df = logins.groupby('time').mean()
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=average_time_df)
timeofdayplot.set_title('Average Logins by Time of Day')
timeofdayplot.set_xticks(range(0,96,4))
timeofdayplot.set_xticklabels([str(x) for x in range(0,24)])
```

```python

day_logins = logins.groupby(['day', 'time']).mean()
day0 = day_logins.loc[0]
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=day0, color='red')

timeofdayplot.set_title('Average Logins by Time of Day')
timeofdayplot.set_xticks(range(0,96,4))
timeofdayplot.set_xticklabels([str(x) for x in range(0,24)]) 

#repeates the above for each day of the week
day1 = day_logins.loc[1]
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=day1, color = 'blue')

day2 = day_logins.loc[2]
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=day2, color = 'green')

day3 = day_logins.loc[3]
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=day3, color = 'purple')

day4 = day_logins.loc[4]
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=day4, color = 'orange')

day5 = day_logins.loc[5]
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=day5, color = 'yellow')

day6 = day_logins.loc[6]
timeofdayplot = sns.lineplot(x=range(0,96), y='count', data=day6, color = 'black')

#labels the lines
```
